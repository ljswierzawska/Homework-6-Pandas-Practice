## 4a) How many male billionaires are there compared to the number of female billionares?

> **TIP:** This is not a `.groupby`

```python
number_female_billionaires = len(df[df['gender'] == 'female'])
number_male_billionaires = len(df[df['gender'] == 'male'])

print(f"There are {number_female_billionaires} female billionaires compared to {number_male_billionaires} male billionaires.")
```

## 4b) What percent is that?

> **TIP:** Doing percentages with the answer to `4a` is so common that there's an extra parameter to automatically calculate it. Look at the documentation!

```python
(df['gender'].value_counts(normalize=True) * 100).round(decimals=2)
```

## 4c) Is the average wealth different for a female billionaire as compared to a male billionaire?

> **TIP:** This one *does* use `.groupby`. You can pick what 'average' means in this context.

```python
average_wealth_female = df[df['gender'] == 'female']['networthusbillion'].median() # I did like this at first, but the task said to use groubpy, so I changed it below. 
average_wealth_male = df[df['gender'] == 'male']['networthusbillion'].median() # I used median instead of mean to account for outliers. 

print(f'The average wealth for a female billionaire is {average_wealth_female} billion US dollars, whereas for male billinaires, it is {average_wealth_male} billion US dollars.')
```

```python
average_wealth_median = df.groupby('gender')['networthusbillion'].median()
average_wealth_median
```

```python
average_wealth_mean = df.groupby('gender')['networthusbillion'].mean()
average_wealth_mean
```

## 5b) Is the most common source/type of wealth different between men and women?

> **TIP:** You know how to `groupby` and you know how to count how many times a value is in a column. Can you put them together??? You could also do this as two separate cells where you calculate it separately for men and women.
>
> **TIP:** Use percentages for this, it makes it a lot more readable.

```python
percentages = (df.groupby('gender')['sourceofwealth'].value_counts(normalize=True) * 100).round(decimals=2)
top_source = percentages.groupby(level=0).head(1)
top_source
```

## 6a) What companies have the most billionaires? Let's see the top five.

## 7) How much money do these top 5 billionaires have in total?

> **TIP:** for multiple conditions, you can't use `or` and `and`. In pandas, `|` means *or* and `&` means *and*.
>
> **TIP:** Checking multiple conditions can be done two ways.
> - `(df.colname == 'A') | (df.colname == 'B')`
> - `df.colname.isin(['A', 'B'])`

```python
top_5_billionaires_total = df["networthusbillion"].sort_values(ascending=False).head(5).sum()

print(f"The top five billionaires have {top_5_billionaires_total} billion US dollars in total.")
```
